```python
s = "leetcode"
k = 3
vowel = {'a', 'e', 'i', 'o', 'u'}
count = 0
max_count = 0 
for i in range(k): 
    if s[i] in vowel:
        count += 1
max_count = count

for right in range(k,len(s)):
    if s[right] in vowel:
        count += 1
    left = right - k
    if s[left] in vowel:
        count -= 1
    max_count = max(max_count,count)
max_count


```
